```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

resim = cv2.imread('picture1.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)  # OpenCV BGR formatında yükler, RGB'ye dönüştürürüz
# Görüntüyü float tipine dönüştürüp normalize etme
resim_float = resim.astype(np.float32) / 255.0
# Logaritma dönüşüm
log_donusum = np.log(1 + resim_float)
# Logaritma dönüşümünü 0-255 aralığına geri dönüştür
log_donusum = cv2.normalize(log_donusum, None, 0, 255, cv2.NORM_MINMAX)
log_donusum = np.uint8(log_donusum)

plt.figure(figsize=(12, 6))
# Orijinal Görüntü
plt.subplot(1, 2, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

# Logaritma Dönüşümlü Görüntü
plt.subplot(1, 2, 2)
plt.imshow(log_donusum, cmap='gray')
plt.title('Logaritma Dönüşümlü Resim')
plt.axis('off')

plt.tight_layout()
plt.show()
```
