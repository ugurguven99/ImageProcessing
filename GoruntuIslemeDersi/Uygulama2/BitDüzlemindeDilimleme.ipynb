```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Görüntüyü gri tonlamalı olarak yükle
resim = cv2.imread('picture1.jpg', 0)

# Görüntü boyutlarını alın
rows, cols = resim.shape

# Bit düzlemlerini saklamak için bir liste oluştur
bit_düzlemleri = []

# 8 bit düzlemine kadar döngü yaparak her bit düzlemini çıkar
for i in range(8):
    # Bit düzlemine göre maske oluştur ve bit düzlemini çıkar
    bit_mask = 1 << i
    bit_plane = cv2.bitwise_and(resim, bit_mask)
    bit_plane = np.where(bit_plane > 0, 255, 0).astype(np.uint8)  # İkili görüntüye dönüştür
    
    bit_düzlemleri.append(bit_plane)

# Görüntüleri göster
plt.figure(figsize=(12, 6))
plt.subplot(3, 4, 1)
plt.imshow(resim, cmap='gray')
plt.title("Orijinal Görüntü")
plt.axis('off')

for i in range(8):
    plt.subplot(3, 4, i+2)
    plt.imshow(bit_düzlemleri[i], cmap='gray')
    plt.title(f'Bit Düzlemi {i}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
