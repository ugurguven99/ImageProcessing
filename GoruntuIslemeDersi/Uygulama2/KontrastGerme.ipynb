```python
#Kontrast germe
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Görüntüyü yükle ve gri tona çevirir
resim = cv2.imread('picture1.jpg', 0)

# Kontrast germe için alt ve üst eşik değerleri
r1, s1 = 70, 0   # Alt sınır (düşük parlaklık)
r2, s2 = 180, 255  # Üst sınır (yüksek parlaklık)

# Kontrast germe fonksiyonu
def kontrast_germe(pixel):
    if pixel < r1:
        return s1
    elif pixel > r2:
        return s2
    else:
        # Lineer bir dönüşüm uygulama
        return ((pixel - r1) * (s2 - s1) / (r2 - r1)) + s1

# Kontrast germe işlemi
vektorize_kontrast_germe = np.vectorize(kontrast_germe)
kontrastli_resim = vektorize_kontrast_germe(resim).astype(np.uint8)

# Orijinal ve kontrast gerilmiş görüntüyü göster
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title("Orijinal Görüntü")
plt.imshow(resim, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Kontrast Gerilmiş Görüntü")
plt.imshow(kontrastli_resim, cmap='gray')
plt.axis('off')
plt.show()
```
