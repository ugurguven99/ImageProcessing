```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

resim = cv2.imread('picture1.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)  # OpenCV BGR formatında yükler, RGB'ye dönüştürürüz

def apply_gamma(image, gamma):
    # Normalize ve gamma dönüşümünü uygula
    gamma_corrected = np.power(image / 255.0, gamma)
    # 0-255 aralığına geri dön ve uint8 tipine çevir
    gamma_corrected = np.uint8(gamma_corrected * 255)
    return gamma_corrected

# Farklı gamma değerleri için dönüşüm uygula
gamma_values = [3.0, 4.0, 5.0]
gamma_images = [apply_gamma(resim, gamma) for gamma in gamma_values]

# Görüntüleri çizdir
plt.figure(figsize=(12, 6))
plt.subplot(1, len(gamma_values) + 1, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

for i, (gamma, gamma_img) in enumerate(zip(gamma_values, gamma_images), start=2):
    plt.subplot(1, len(gamma_values) + 1, i)
    plt.imshow(gamma_img)
    plt.title(f'Gamma: {gamma}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
