```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

resim = cv2.imread('picture3.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)

def add_exponential_noise(image, scale):
    # Üstel dağılımından rastgele değerler oluştur
    noise = np.random.exponential(scale, image.shape).astype(np.uint8)
    # Gürültüyü görüntüye ekle
    noisy_image = cv2.add(image, noise)
    return noisy_image
scale = 25.0  # Üstel dağılımı ölçek parametresi
noisy_resim = add_exponential_noise(resim, scale)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(noisy_resim)
plt.title('Üstel Gürültülü Resim')
plt.axis('off')

plt.tight_layout()
plt.show()
```
