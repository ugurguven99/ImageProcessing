```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


resim = cv2.imread('picture3.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)

# Gauss gürültüsü eklemek için bir fonksiyon tanımla
def add_gaussian_noise(image, mean=0, sigma=25):
    # Gürültüyü oluştur
    gaussian = np.random.normal(mean, sigma, image.shape).astype(np.uint8)
    # Gürültüyü görüntüye ekle
    noisy_image = cv2.add(image, gaussian)
    return noisy_image

# Gauss gürültüsü ekle
noisy_resim = add_gaussian_noise(resim, mean=0, sigma=25)

# Görüntülerin Görselleştirilmesi
plt.figure(figsize=(12, 6))

# Orijinal Görüntü
plt.subplot(1, 2, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

# Gürültülü Görüntü
plt.subplot(1, 2, 2)
plt.imshow(noisy_resim)
plt.title('Gauss Gürültülü Resim')
plt.axis('off')

plt.tight_layout()
plt.show()
```
