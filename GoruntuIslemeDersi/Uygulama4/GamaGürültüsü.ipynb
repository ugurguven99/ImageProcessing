```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


resim = cv2.imread('picture3.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)

def add_gamma_noise(image, shape, scale):
    
    noise = np.random.gamma(shape, scale, image.shape).astype(np.uint8)
    
    noisy_image = cv2.add(image, noise)
    return noisy_image


shape = 2.0  
scale = 10.0  
noisy_resim = add_gamma_noise(resim, shape, scale)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(noisy_resim)
plt.title('Gamma G端r端lt端l端 Resim')
plt.axis('off')

plt.tight_layout()
plt.show()
```
