```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Görüntüyü yükle
resim = cv2.imread('picture1.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)

# Laplace Operatörünü tanımlama
laplace_kernel = np.array([[0, 1, 0],
                            [1, -4, 1],
                            [0, 1, 0]])

# Görüntüye Laplace Operatörünü uygulama
laplacian_resim = cv2.filter2D(resim, -1, laplace_kernel)

# Görüntüyü keskinleştirmek için orijinal görüntüye ekleme
keskinlesmis_resim = cv2.add(resim, laplacian_resim)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(laplacian_resim)
plt.title('Laplace Operatörü Uygulanmış Resim')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(keskinlesmis_resim)
plt.title('Keskinleştirilmiş Resim')
plt.axis('off')

plt.tight_layout()
plt.show()
```
