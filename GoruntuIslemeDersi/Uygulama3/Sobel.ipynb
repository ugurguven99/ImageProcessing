```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Görüntüyü yükle
resim = cv2.imread('picture1.jpg')
resim_gray = cv2.cvtColor(resim, cv2.COLOR_BGR2GRAY)  # Görüntüyü gri tonlamalıya çevir

# Sobel Operatörünü uygula (X ve Y yönünde)
sobel_x = cv2.Sobel(resim_gray, cv2.CV_64F, 1, 0, ksize=3)  # X yönünde
sobel_y = cv2.Sobel(resim_gray, cv2.CV_64F, 0, 1, ksize=3)  # Y yönünde

# Kenarları birleştirme (magnitude)
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)

# Sonucu normalize et ve uint8 formatına çevir
sobel_magnitude = cv2.normalize(sobel_magnitude, None, 0, 255, cv2.NORM_MINMAX)
sobel_magnitude = np.uint8(sobel_magnitude)


plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

# Sobel X Yönünde
plt.subplot(1, 3, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X Yönü')
plt.axis('off')

# Sobel Y Yönünde
plt.subplot(1, 3, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y Yönü')
plt.axis('off')

plt.tight_layout()
plt.show()

# Kenarların büyüklüğünü göster
plt.figure(figsize=(6, 6))
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Sobel Kenar Magnitude')
plt.axis('off')
plt.show()
```
