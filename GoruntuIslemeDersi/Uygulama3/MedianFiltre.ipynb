```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

resim = cv2.imread('picture1.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)

# Medyan Filtresi uygula
kernel_size = 5  # Filtresi boyutu (tek sayı olmalı)
yumusatilmis_resim = cv2.medianBlur(resim, kernel_size)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(yumusatilmis_resim)
plt.title('Medyan Filtresi Uygulanmış Resim')
plt.axis('off')

plt.tight_layout()
plt.show()
```
