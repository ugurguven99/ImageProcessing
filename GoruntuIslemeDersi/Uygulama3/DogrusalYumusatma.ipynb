```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

resim = cv2.imread('picture1.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)

# Doğrusal Yumuşatma Filtresi (Ortalama Filtresi) uygula
kernel_size = (5, 5)  # Filtresi boyutu
yumusatilmis_resim = cv2.blur(resim, kernel_size)

plt.figure(figsize=(12, 6))

# Orijinal Görüntü
plt.subplot(1, 2, 1)
plt.imshow(resim)
plt.title('Orijinal Resim')
plt.axis('off')

# Yumuşatılmış Görüntü
plt.subplot(1, 2, 2)
plt.imshow(yumusatilmis_resim)
plt.title('Yumuşatılmış Resim')
plt.axis('off')

plt.tight_layout()
plt.show()
```
