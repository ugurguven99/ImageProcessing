```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

resim = cv2.imread('picture1.jpg')
resim_gray = cv2.cvtColor(resim, cv2.COLOR_BGR2GRAY)  

# Roberts Çapraz Gradyan Kernel'leri
roberts_x = np.array([[1, 0],
                       [0, -1]])

roberts_y = np.array([[0, 1],
                       [-1, 0]])

# Görüntüye Roberts Operatörünü uygula
roberts_grad_x = cv2.filter2D(resim_gray, cv2.CV_64F, roberts_x)
roberts_grad_y = cv2.filter2D(resim_gray, cv2.CV_64F, roberts_y)

# Kenar büyüklüğünü hesapla
roberts_magnitude = cv2.magnitude(roberts_grad_x, roberts_grad_y)

# Sonucu normalize et ve uint8 formatına çevir
roberts_magnitude = cv2.normalize(roberts_magnitude, None, 0, 255, cv2.NORM_MINMAX)
roberts_magnitude = np.uint8(roberts_magnitude)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(resim)

plt.title('Orijinal Resim')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(roberts_grad_x, cmap='gray')
plt.title('Roberts X Yönü')
plt.axis('off')

# Roberts Y Yönünde
plt.subplot(1, 3, 3)
plt.imshow(roberts_grad_y, cmap='gray')
plt.title('Roberts Y Yönü')
plt.axis('off')

plt.tight_layout()
plt.show()

# Kenar büyüklüğünü göster
plt.figure(figsize=(6, 6))
plt.imshow(roberts_magnitude, cmap='gray')
plt.title('Roberts Kenar Magnitude')
plt.axis('off')
plt.show()
```
